```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.util import ngrams
stop = stopwords.words('english')
```

```python
review_path = "data/yelp_review_training_dataset.jsonl"
df = pd.read_json(review_path, lines = True)
```

```python
df.head()
```

```python
df = df.drop(["review_id"], axis = 1)
df
```

```python
df["text"] = df['text'].str.replace('[^\w\s]','').str.lower()
df["text"] = df['text'].str.replace('\d+', '')
```

```python
df
```

```python
stop
```

```python


df['text'] = df['text'].apply(lambda x: [item for item in x.split() if item not in stop])
```

```python
df
```

```python
df.to_csv('clean1gram_yelp_review_training_dataset.csv', header=False, index=False) 
```
